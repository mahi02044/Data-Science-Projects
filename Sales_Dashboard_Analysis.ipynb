```python
# These libraries come pre-installed in Google Colab
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
print("✅ Libraries loaded successfully!")
```

```python
# Load the CSV file
df = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")

print("✅ Dataset loaded!")
print("Shape:", df.shape)  # Shows (rows, columns)
df.head()                  # Preview first 5 rows
```

```python
# Check column names
print(df.columns.tolist())
```

```python
# Check data types and missing values
df.info()
```

```python
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert Order Date to proper date format
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Extract Year and Month
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()

print("✅ Data cleaned!")
df.head()
```

```python
plt.figure(figsize=(8, 5))
category_sales = df.groupby("Category")["Sales"].sum().sort_values()
category_sales.plot(kind="barh", color=["#4C72B0", "#DD8452", "#55A868"])
plt.title("Total Sales by Category", fontsize=16)
plt.xlabel("Sales ($)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
region_profit = df.groupby("Region")["Profit"].sum()
plt.pie(region_profit, labels=region_profit.index,
        autopct="%1.1f%%", startangle=140)
plt.title("Profit Distribution by Region", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
monthly_sales = df.resample("M", on="Order Date")["Sales"].sum()
monthly_sales.plot(color="steelblue", linewidth=2)
plt.title("Monthly Sales Trend Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales ($)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_products = df.groupby("Product Name")["Profit"].sum()\
                 .sort_values(ascending=False).head(10)
top_products.plot(kind="bar", color="seagreen")
plt.title("Top 10 Most Profitable Products", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Profit ($)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Sales", y="Profit",
                hue="Category", alpha=0.6)
plt.title("Sales vs Profit by Category", fontsize=16)
plt.axhline(0, color="red", linestyle="--", label="Break-even line")
plt.legend()
plt.tight_layout()
plt.show()
```
